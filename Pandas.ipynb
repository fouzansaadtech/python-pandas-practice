```python
import pandas as pd
i=pd.read_excel("Pandas_Practice_Dataset.xlsx")
```

```python
# What is the total sales generated

n=i["Sales"].sum()
print("Total Sales =",n)
```

```python
#Which are the top 5 customers based on total sales?

Top5_cus=i.groupby(["Customer"])["Sales"].sum().sort_values(ascending=False).head(5)
print("Top 5 customers based on total sales:",Top5_cus)
```

```python
# Which are the top 5 products based on total sales?

Top5_pro=i.groupby(["Product"])["Sales"].sum().sort_values(ascending=False).head(5)
print("Top 5 products based on total sales:",Top5_pro)
```

```python
#Which category generated the highest total sales?

cate=i.groupby(["Category"])["Sales"].sum().sort_values(ascending=False).head(1)
print("Category generated the highest total sales:",cate)

```

```python
# Which city generated the highest total sales?

city=i.groupby(["City"])["Sales"].sum().sort_values(ascending=False).head(1)
print("City generated the highest total sales",city)
```

```python
#Which payment method was used the most

payment=i.value_counts(["Payment"]).sort_values(ascending=False).head(1)
print("Payment method was used the most",payment)
```

```python
#Find all orders where Sales > 5000 and Profit > 1000.

Sales=i[(i["Sales"]>5000) & (i["Profit"]>1000)]
print("Sales > 5000 and Profit > 1000:",Sales)
```

```python
#Create a pivot table showing total sales by City and Category.

i.pivot_table(values="Sales", index="City", columns="Category", aggfunc="sum")
```

```python
#Which customer spent the most on Furniture products

i[i["Category"] == "Furniture"] \
    .groupby("Customer")["Sales"] \
    .sum() \
    .sort_values(ascending=False) \
    .head(1)
```

```python
#Export the top 10 customers by sales into a new Excel file named Top_Customers.xlsx.

export=i.groupby(["Customer"])["Sales"].sum().sort_values(ascending=False).head(10)
export.to_excel("Top_Customers1.xlsx")
```
